```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model

np.random.seed(4)
#Hàm rand tạo số ngẫu nhiên có phân phối đều nằm trong khoảng (0,1)
#Ví dụ: cho N = 30
#       5*rand(N,1)tạo số ma trận gồm N hàng x 1 cột và có giá trị từ 0 đến 5
# ham ramdn (n = normal gaussion : chuẩn ) taoj so ngua nhien co phan phoi chuan (gauss) co mean = 0 va standand deration = 1 ; 
# ký hiêu trung bình là 
# ký hiêu độ lệch chuẩn là 
N = 30
#Tạo 30 số ngẫu nhiên
#phân phối chuẩn từ  0,5
X= np.random.rand(N,1)*5
#Tạo N số ngẫu nhiên có
#phân phối chuẩn với mean =0
# và độ lệch chuẩn =10
noise =10*np.random.randn(N,1)
y = 3*(X-2)*(X-3)*(X-4)+ noise

poly_features = PolynomialFeatures(degree=3, include_bias=False)
X_poly = poly_features.fit_transform(X)

#print(X[0])
#print(X)

lin_reg = linear_model.LinearRegression()
lin_reg.fit(X_poly, y)

print(lin_reg.intercept_)
print(lin_reg.coef_)
#w0 = lin_reg.intercept_[0]
#w1 = lin_reg.coef_[0,0]
#w2 = lin_reg.coef_[0,1]
#w3 = lin_reg.coef_[0,2]
X_true = np.linspace(-1,5.5,21)# mảng một chiều 21 số:(21,) 
X_true = np.expand_dims(X_true,axis =1)
y_true = 3*(X_true-2)*(X_true-3)*(X_true-4)

X_true_poly= poly_features.fit_transform(X_true)
y_mu =lin_reg.predict(X_true_poly)

#pass#lệnh không làm gì cả
#y_mu=w0+w1*X_true+ w2*X_true**2+ w3*X_true**3

plt.plot(X_true,y_true,'--g')
plt.plot(X, y, 'ro')
plt.plot(X_true,y_mu,'b')

plt.xlim(-3,8)
plt.ylim(-130,50)
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model

np.random.seed(4)
N_test = 30
N = 30
#Tạo 30 số ngẫu nhiên
#phân phối chuẩn từ  0,5
X= np.random.rand(N,1)*5
#Tạo N số ngẫu nhiên có
#phân phối chuẩn với mean =0
# và độ lệch chuẩn =10
noise =10*np.random.randn(N,1)
y = 3*(X-2)*(X-3)*(X-4)+ noise

X_test = (np.random.rand(N_test,1)-1/8)*10
y_test =  3*(X_test-2)*(X_test-3)*(X_test-4)+ 10*np.random.randn(N_test,1)
bac =0 

def vedathuc(bac):  
    poly_features = PolynomialFeatures(degree=bac, include_bias=False)
    X_poly = poly_features.fit_transform(X)

    lin_reg = linear_model.LinearRegression()
    lin_reg.fit(X_poly, y)



    X_true = np.linspace(-1.5,6.5,51)# mảng một chiều 21 số:(21,) 
    X_true = np.expand_dims(X_true,axis =1)
    y_true = 3*(X_true-2)*(X_true-3)*(X_true-4)

    X_true_poly= poly_features.fit_transform(X_true)
    y_mu =lin_reg.predict(X_true_poly)

    if bac ==2 : 
        plt.subplot(2,2,1)
    elif bac == 4: 
        plt.subplot(2,2,2)
    elif bac == 6: 
        plt.subplot(2,2,3)
    elif bac == 8: 
        plt.subplot(2,2,4)



    plt.plot(X,y,'ro')
    plt.plot(X_test,y_test,'ys')  # y : yellow , s : square 
    plt.plot(X_true,y_true,'black')
    plt.plot(X_true,y_mu,'b')



    plt.xlim(-4,7)
    plt.ylim(-120,150)
    plt.title('Da thuc bac %d'  % bac )

  
vedathuc(2)
vedathuc(4)
vedathuc(6)
vedathuc(8)
plt.tight_layout()
plt.show()
```

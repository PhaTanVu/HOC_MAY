```python
import pandas as pd
housing = pd.read_csv('/content/housing1.csv')
print(housing.tail())
print(housing.info())
housing["ocean_proximity"].value_counts()
housing.describe()
print(housing.describe())

vi_tri = []
for i in range(len(housing)):
    vi_tri.append(housing.iloc[i, 9])
vi_tri = sorted(vi_tri)

key = '*'
dem = 0
for x in vi_tri:
    if (key != x):
        if key != '*':
            print(key, dem)
        key = x
        dem = 0
    dem = dem + 1
print(key, dem)
sum = 0
for i in range(len(housing)):
    tuoi_ngoi_nha = housing.iloc[i, 2]
    sum = sum + tuoi_ngoi_nha
trung_binh = sum/len(housing)
print('%.6f' % trung_binh)

import numpy as np

tong = 0
for i in range(len(housing)):
    tuoi_ngoi_nha = housing.iloc[i, 2]
    tong = tong + ( tuoi_ngoi_nha - trung_binh)**2
phuong_sai = tong/(len(housing)-1)

do_lech_chuan = np.sqrt(phuong_sai)
print('DO LECH CHUAN', do_lech_chuan)


tuoi_ngoi_nha =[]
for i in range(len(housing)):
    tuoi_ngoi_nha.append(housing.iloc[i, 2])

tuoi_ngoi_nha = sorted(tuoi_ngoi_nha)
tuoi_ngoi_nha = np.array(tuoi_ngoi_nha)
min = np.min(tuoi_ngoi_nha)
max = np.max(tuoi_ngoi_nha)

print('MIN', min)
print('MAX', max)

vi_tri_25 = 25*len(housing) // 100
print(tuoi_ngoi_nha[vi_tri_25 -1])

vi_tri_50 = 50*len(housing) // 100
print(tuoi_ngoi_nha[vi_tri_50 -1])

vi_tri_75 = 75*len(housing) // 100
print(tuoi_ngoi_nha[vi_tri_75 -1])
```
